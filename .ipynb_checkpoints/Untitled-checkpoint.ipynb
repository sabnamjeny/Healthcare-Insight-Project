```python
# ডেটাসেট লোড করা
df = pd.read_excel("C:/Users/Hp/OneDrive/Desktop/Healthcare Insides Dasbord project 1.xlsx")

# ১. Followup Date কলাম বাদ দেওয়া
df = df.drop(columns=['Followup Date'])

# ২. Length_of_Stay নতুন কলাম তৈরি করা
df['Length_of_Stay'] = (df['Discharge_Date'] - df['Admit_Date']).dt.days

# ৩. তৈরি হওয়া ডেটাসেট দেখে নেওয়া
df.head()
```

```python
# লাইব্রেরি ইম্পোর্ট
import sqlite3

# ১. ডাটাবেজ ফাইল তৈরি করা (healthcare.db নামে)
conn = sqlite3.connect('healthcare.db')

# ২. ডেটাসেট healthcare টেবিলে আপলোড করা
df.to_sql('healthcare', conn, if_exists='replace', index=False)

# ৩. ডেটা দেখতে চাইলে
query_result = pd.read_sql_query("SELECT * FROM healthcare LIMIT 5;", conn)
print(query_result)

# ৪. কাজ শেষে কানেকশন বন্ধ করা
conn.close()
```

```python


# যদি Length_of_Stay নাই থাকে, তাহলে এটা আগে যোগ করে নিতে হবে
df['Admit_Date'] = pd.to_datetime(df['Admit_Date'])
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'])
df['Length_of_Stay'] = (df['Discharge_Date'] - df['Admit_Date']).dt.days

# 3. ডাটাবেজ কানেকশন তৈরি করি
conn = sqlite3.connect('healthcare.db')

# 4. healthcare নামে একটা টেবিল বানিয়ে ডেটা ঢোকাই
df.to_sql('healthcare', conn, if_exists='replace', index=False)

# 5. ভেরিফাই করি — দেখি টেবিল বানানো হয়েছে কিনা
query_result = pd.read_sql_query("SELECT * FROM healthcare LIMIT 5;", conn)
print(query_result)

# 6. কানেকশন বন্ধ করে দেই
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('healthcare.db')

# SQL query to analyze admission trends (use correct column names)
query = """
SELECT strftime('%Y-%m', Admit_Date) AS admission_month, COUNT(*) AS total_admissions
FROM healthcare_data
GROUP BY admission_month
ORDER BY admission_month;
"""

# Execute the query and load the result into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Display the results
print(df)

# Close the connection
conn.close()
```

```python
from pandasql import sqldf
import pandas as pd

# healthcare ডেটাফ্রেম তৈরি করা (তুমি অবশ্যই এই ধরণের ডেটাফ্রেমের সাথে কাজ করবে)
healthcare_data = pd.DataFrame({
    'Patient_ID': [23571, 27781, 24413, 27360, 26097],
    'Admit_Date': ['2022-12-31', '2023-01-04', '2023-01-05', '2023-01-05', '2023-01-06'],
    'Discharge_Date': ['2023-01-12', '2023-01-12', '2023-01-12', '2023-01-12', '2023-01-12'],
    'Diagnosis': ['Viral Infection', 'Typhoid', 'Malaria', 'Flu', 'Viral Infection'],
    'Bed_Occupancy': ['General', 'ICU', 'General', 'Private', 'General'],
    'Test': ['MRI', 'MRI', 'CT Scan', 'X-Ray', 'Blood Test'],
    'Doctor': ['Jay Sinha', 'Jaya Yaadav', 'Jay Sinha', 'Jaya Yaadav', 'Jaya Yaadav'],
    'Feedback': [5.0, 4.0, 5.0, 5.0, 4.9],
    'Billing Amount': [13234, 12241, 35255, 6582, 4567],
    'Health Insurance Amount': [11910.6, 11016.9, 31729.5, 5923.8, 4110.3],
    'Length_of_Stay': [12, 8, 7, 7, 6]
})

# pandasql কোয়েরি সেটআপ
pysqldf = lambda q: sqldf(q, globals())

# SQL কোড
query = """
SELECT 
    STRFTIME('%Y-%m', Admit_Date) AS Month, 
    COUNT(Patient_ID) AS Admissions
FROM healthcare_data
GROUP BY Month
ORDER BY Month;
"""

# SQL কোড চালানো
result = pysqldf(query)
print(result)
```
